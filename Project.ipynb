```python
import requests
import pandas as pd
from datetime import datetime

# Set your API key
api_key = '914cea15e39cfad83e282b6b800068b7'

# Fetch Historical Stock Data
stock_symbol = 'AAPL'  # Apple Inc.
url_stock = f'https://financialmodelingprep.com/api/v3/historical-price-full/{stock_symbol}?apikey={api_key}'

# Fetch Market Index and Ratios
url_market_index = f'https://financialmodelingprep.com/api/v3/market-index/{stock_symbol}?apikey={api_key}'
url_ratios = f'https://financialmodelingprep.com/api/v3/ratios/{stock_symbol}?apikey={api_key}'

# Function to make API request and return data
def get_data(url):
    response = requests.get(url)
    if response.status_code == 200:
        return response.json()
    else:
        print(f'Failed to retrieve data: {response.status_code}')
        return None

# Fetch data
stock_data = get_data(url_stock)
market_index_data = get_data(url_market_index)
ratios_data = get_data(url_ratios)

# Process and display historical stock data
if stock_data and 'historical' in stock_data:
    df_stock = pd.DataFrame(stock_data['historical'])
    df_stock['date'] = pd.to_datetime(df_stock['date'])
    ten_years_ago = datetime.now() - pd.DateOffset(years=10)
    df_stock_last_10_years = df_stock[df_stock['date'] >= ten_years_ago]
    print(df_stock_last_10_years.head())
else:
    print("No historical stock data found.")

# Process and display market index data
if market_index_data:
    df_market_index = pd.DataFrame(market_index_data)
    print(df_market_index.head())

# Process and display ratios data
if ratios_data:
    df_ratios = pd.DataFrame(ratios_data)
    print(df_ratios.head())
```

```python
import requests

def get_eurostat_data(dataset_code, year, format='json', lang='EN'):
    """
    Fetches data from the Eurostat API.

    :param dataset_code: Code of the Eurostat dataset.
    :param year: Year for which data is requested.
    :param format: Format of the response (json or unicode).
    :param lang: Language of the response.
    :return: Data in the requested format.
    """
    base_url = "https://ec.europa.eu/eurostat/api/dissemination/statistics/1.0/data/"
    url = f"{base_url}{dataset_code}?format={format}&lang={lang}&time={year}"

    response = requests.get(url)
    if response.status_code == 200:
        return response.json() if format == 'json' else response.text
    else:
        return f"Error: Failed to retrieve data with status code {response.status_code}"

# Example usage
dataset_code = 'irt_euryld_d'  # Eurostat dataset code for interest rates or yields
year = 2019  # Example year
data = get_eurostat_data(dataset_code, year)
print(data)
```

```python
import requests
import gzip
import shutil

# Dataset code
dataset_code = "irt_euryld_d"

# API URL for downloading the dataset
api_url = f"https://ec.europa.eu/eurostat/api/dissemination/sdmx/2.1/data/{dataset_code}?format=TSV&compressed=true"

# Download the dataset
response = requests.get(api_url, stream=True)
output_file = dataset_code + ".tsv.gz"

if response.status_code == 200:
    with open(output_file, 'wb') as f:
        f.write(response.content)
    print("Dataset downloaded successfully.")
else:
    print(f"Error downloading dataset. Status code: {response.status_code}")

# Unzipping the dataset
with gzip.open(output_file, 'rb') as f_in:
    with open(dataset_code + ".tsv", 'wb') as f_out:
        shutil.copyfileobj(f_in, f_out)

print("Dataset unzipped successfully.")
```

```python
import pandas as pd

# Load the dataset (replace the file path with your dataset's file path)
file_path = '/home/fabiopoli/Desktop/irt_euryld_d.tsv'
data = pd.read_csv(file_path, sep='\t')

# Split the first column into separate columns
split_columns = data.iloc[:, 0].str.split(',', expand=True)
split_columns.columns = ['freq', 'yld_curv', 'maturity', 'bonds', 'geo']

# Combine the split columns with the original data (excluding the first column)
structured_data = pd.concat([split_columns, data.iloc[:, 1:]], axis=1)

# Apply filters
structured_data = structured_data[
    (structured_data['yld_curv'] == 'SPOT_RT') & 
    (structured_data['bonds'] == 'CGB_EA')
]

# Drop unnecessary columns
structured_data.drop(['freq', 'geo'], axis=1, inplace=True)

# Filter for date range (from October 1, 2004, onwards)
date_columns = structured_data.columns[3:]
dates_from_oct_2004 = date_columns[date_columns >= '2004-10-01 ']
structured_data = structured_data[['yld_curv', 'maturity', 'bonds'] + list(dates_from_oct_2004)]

# Reset the index for a clean table
structured_data.reset_index(drop=True, inplace=True)

# Display the structured DataFrame
pd.set_option('display.max_columns', None)
print(structured_data.head(10))
```
